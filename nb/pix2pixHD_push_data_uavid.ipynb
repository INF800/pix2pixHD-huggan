```python
""" create dataset objects for splits """

from pathlib import Path
from datasets import load_dataset, Image, Dataset, concatenate_datasets
from pprint import pprint

DATA_DIR = Path("/content/input/uavid15hqmixeddims/uavid_v1.5_without_video")

train_images = sorted(list((DATA_DIR / 'uavid_train').glob("seq*/Images/*.png")))
train_labels = sorted(list((DATA_DIR / 'uavid_train').glob("seq*/Labels/*.png")))

test_images = sorted(list((DATA_DIR / 'uavid_val').glob("seq*/Images/*.png")))
test_labels = sorted(list((DATA_DIR / 'uavid_val').glob("seq*/Labels/*.png")))

train_images = [str(i) for i in train_images]
train_labels = [str(i) for i in train_labels]
test_images = [str(i) for i in test_images]
test_labels = [str(i) for i in test_labels]


print("train: ", len(train_images), len(train_labels))
print("valid: ", len(test_images), len(test_labels))


train_ds  = Dataset.from_dict({
  "image": train_images,
  "label": train_labels,
}).cast_column("image", Image()).cast_column("label", Image())#.cast_column("instance_id", Image())

test_ds  = Dataset.from_dict({
  "image": test_images,
  "label": test_labels,
}).cast_column("image", Image()).cast_column("label", Image())# .cast_column("instance_id", Image())


pprint(train_ds[1])
pprint(test_ds[1])
```

```python
"""push the splits to hub"""
import datasets

ds = datasets.DatasetDict(train=train_ds, test=test_ds)
ds.push_to_hub("uavid-15-hq-mixedres")
```
